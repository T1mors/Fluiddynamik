```python
import numpy as np
from matplotlib import pyplot as plt
import math
import seaborn as sns
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
plt.rc('font', family='serif')
#plt.style.use('seaborn-white',)
#rc('text', usetex=True)     #this CAN cause crashes, solution: cd ~/.matplotlib  then  rm tex.cache/
sns.set_style("whitegrid")
```

```python
def radial(t, mu, r_i):
    return np.sqrt(mu)*(1.+(mu/r_i**2-1.)*np.exp(-2.*mu*t))**(-1/2)
def radial_neg_mu(t, mu, r_i):
    return np.sqrt(-mu)*(-1.-(mu/r_i**2-1.)*np.exp(-2.*mu*t))**(-1/2)
```

```python
mu = 2; r_i = 3.; beta_i = 1.; T = 1.; omega = 2*np.pi/T
R = np.vectorize(radial)
t = np.linspace(0,3*T,2000)
phi = omega*t + beta_i 
x = R(t, mu, r_i)*np.cos(phi)
y = R(t, mu, r_i)*np.sin(phi)
plt.plot(x, y)
plt.scatter(x[0], y[0])
phi = omega*t - 1. 
x = R(t, mu, 0.2)*np.cos(phi)
y = R(t, mu, 0.2)*np.sin(phi)
plt.plot(x, y)
plt.scatter(x[0], y[0])
plt.ylabel('$y$', fontsize=20)
plt.xlabel('$x$', fontsize=20)
#plt.legend(fontsize=12)
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)
#plt.yscale('log')
plt.savefig('hopf_mu_2.pdf', bbox_inches='tight')
```

```python
mu = -2.; r_i = 3.; beta_i = 1.; T = 1.; omega = 2*np.pi/T
R = np.vectorize(radial_neg_mu)
t = np.linspace(0,3*T,2000)
phi = omega*t + beta_i 
x = R(t, mu, r_i)*np.cos(phi)
y = R(t, mu, r_i)*np.sin(phi)
plt.plot(x, y)
plt.scatter(x[0], y[0])
phi = omega*t - 1. 
x = R(t, mu, 0.2)*np.cos(phi)
y = R(t, mu, 0.2)*np.sin(phi)
plt.plot(x, y)
plt.scatter(x[0], y[0])
plt.ylabel('$y$', fontsize=20)
plt.xlabel('$x$', fontsize=20)
#plt.legend(fontsize=12)
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)
#plt.yscale('log')
plt.savefig('hopf_mu_neg_2.pdf', bbox_inches='tight')
```
