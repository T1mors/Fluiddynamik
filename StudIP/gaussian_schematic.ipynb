```python
import numpy as np
from matplotlib import pyplot as plt
import math
import seaborn as sns
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
plt.rc('font', family='serif')
#plt.style.use('seaborn-white',)
#rc('text', usetex=True)     #this CAN cause crashes, solution: cd ~/.matplotlib  then  rm tex.cache/
sns.set_style("whitegrid")
```

```python
def gaussian(u, sigma, U):
    return np.exp(-(u-U)**2/(2*sigma**2))/np.sqrt(2*np.pi*sigma**2)
```

```python
sigma = 1; U = 2*sigma
u = np.linspace(-6*sigma, 8*sigma, 300)
Gaussian = np.vectorize(gaussian)
plt.plot(u, Gaussian(u, sigma, U), label='$\\sigma=1$, $U=1$')
sigma = 1.; U = 0*sigma
plt.plot(u, Gaussian(u, sigma, U), label='$\\sigma=1$, $U=0$')
sigma = 2; U = 0*sigma
plt.plot(u, Gaussian(u, sigma, U), label='$\\sigma=2$, $U=0$')
plt.ylabel('$g(u)$', fontsize=20)
plt.xlabel('$u$', fontsize=20)
plt.legend(fontsize=12)
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)
#plt.yscale('log')
plt.savefig('gaussians_schematic.pdf', bbox_inches='tight')
```
